```python
import cv2
from PIL import Image
def read_image(path):
    img = cv2.imread(str(path))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img
images=[]



```

```python
df['filename'][0]
```

```python
import os
read_image(os.path.join('.','train',df['filename'][1]))
```

```python
for i in df['filename']:
    img= read_image(os.path.join('.','train',i))
    images.append(img)
```

```python
images=np.array(images)
```

```python
images.shape
```

```python
y=df['class']
y
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y= le.fit_transform(y)
```

```python
y=pd.Series(y)
```

```python
y.value_counts()
```

```python
cnn= keras.models.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(640, 640, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(128, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),

    keras.layers.Dense(32, activation='relu'),
    

    keras.layers.Dense(len(y.unique()), activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(images,y,epochs=3)
```
